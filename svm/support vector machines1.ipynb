```python
categorical = df.select_dtypes(include=['object'])

categorical.head()
```

```python
numerical = df.select_dtypes(include=['int64', 'float64'])

numerical.head()
```

```python
numerical = numerical.drop(['Salary'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
df_num = pd.DataFrame(scaler.fit_transform(numerical), 
                      columns = numerical.columns)

df_num.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='first', sparse=False)

df_cat = pd.DataFrame(encoder.fit_transform(categorical), 
                      columns=encoder.get_feature_names(categorical.columns))

df_cat.head()
```

```python
df1 = pd.concat([df_num, df_cat], axis=1)

df1.head()
```

```python
x = df1
y= df['Salary']
```

```python
test.head()
```

```python
numerical_test = test.select_dtypes(include=['int64', 'float64'])

numerical_test.head()
```

```python
numerical_test = numerical_test.drop(['Salary'],axis=1)
```

```python
test_num = pd.DataFrame(scaler.transform(numerical_test), 
                      columns = numerical_test.columns)

test_num.head()
```

```python
categorical_test = test.select_dtypes(include=['object'])

categorical_test.head()
```

```python
test_cat = pd.DataFrame(encoder.transform(categorical_test), 
                      columns=encoder.get_feature_names(categorical_test.columns))

test_cat.head()
```

```python
test1 = pd.concat([test_num, test_cat], axis=1)

test1.head()
```

```python
from sklearn.svm import SVC
```

```python
#Linear
```

```python
from sklearn import metrics
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x,y)
pred_test_linear = model_linear.predict(test1)
print("Accuracy:",metrics.accuracy_score(test['Salary'], pred_test_linear))
```

```python
#pOLY
```

```python
model_poly = SVC(kernel = "poly")
model_poly.fit(x,y)
pred_test_poly = model_poly.predict(test1)
print("Accuracy:",metrics.accuracy_score(test['Salary'], pred_test_poly))
```

```python
#rbf
```

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x,y)
pred_test_rbf = model_rbf.predict(test1)
print("Accuracy:",metrics.accuracy_score(test['Salary'], pred_test_rbf))
```

```python
#All different kernals yeilded similar accuracies of 84% of which rbf kernal performed slight better
```
